```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('processed-data/my_book_post.csv')
df.head()
```

```python
df['topic'].unique()
```

```python
minimal_df = df.groupby('topic').apply(lambda x: x.head(500)).reset_index(drop=True)
minimal_df.groupby('topic').count()
```

```python
import json

result = minimal_df.to_json(orient='records')
parsed = json.loads(result)
json_out = json.dumps(parsed, indent=4)

with open('processed-data/my_book_post_minimal.json', 'w') as file:
    file.write(json_out)
```
